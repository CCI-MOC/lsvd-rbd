```python
import os
import re
import csv
from datetime import datetime
import smtplib
from email.mime.text import MIMEText
import subprocess
from subprocess import check_output
import matplotlib.pyplot as plt
import pandas as pd
import time
import hashlib
import numpy as np

def extract_values(input_list):
    numeric_values = []
    converted_values = []

    for value_with_unit in input_list:
        match = re.match(r'(\d+\.\d+|\d+)(\w+)/s', value_with_unit)
        if match:
            numeric_value = float(match.group(1))
            unit = match.group(2)

            if unit == 'kb/s':
                numeric_value /= 1024  # Convert KiB to MiB

            numeric_values.append(numeric_value)
            converted_values.append(f'{numeric_value}mb/s')

    return numeric_values

def condition_generator(cs, pt, dt):
    conditions_lsvd_ssd_fs = (
        (df['pool_type'] == pt) & 
        (df['disk_type'] == dt) & 
        (df['cache_size'] == cs) & 
        (df['file_type'] == 'fileserver')
    )
    
    conditions_lsvd_ssd_oltp = (
        (df['pool_type'] == pt) & 
        (df['disk_type'] == dt) & 
        (df['cache_size'] == cs) & 
        (df['file_type'] == 'oltp') 
    )
    
    conditions_lsvd_ssd_vm = (
        (df['pool_type'] == pt) & 
        (df['disk_type'] == dt) & 
        (df['cache_size'] == cs) & 
        (df['file_type'] == 'varmail')
    )
    
    return conditions_lsvd_ssd_fs, conditions_lsvd_ssd_oltp, conditions_lsvd_ssd_vm
```
