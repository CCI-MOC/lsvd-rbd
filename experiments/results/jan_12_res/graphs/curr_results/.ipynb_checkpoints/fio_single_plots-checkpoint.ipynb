```python
df = pd.read_csv("single_fio.csv")

bs='4ki'
qd=256
cs='20gb'
pt='ssd'
dt='lsvd'

#condition for lsvd_ssd
conditions_lsvd_ssd_rr = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'randread') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

conditions_lsvd_ssd_rw = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'randwrite') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

conditions_lsvd_ssd_sr = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'read') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

conditions_lsvd_ssd_sw = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'write') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)




pt ='hdd'
#confition for lsvd_hdd
conditions_lsvd_hdd_rr = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'randread') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

conditions_lsvd_hdd_rw = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'randwrite') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

conditions_lsvd_hdd_sr = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'read') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

conditions_lsvd_hdd_sw = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'write') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)




dt='rbd'
pt='ssd'
cs='none'
#condition for rbd_ssd
conditions_rbd_ssd_rr = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'randread') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

conditions_rbd_ssd_rw = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'randwrite') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

conditions_rbd_ssd_sr = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'read') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

conditions_rbd_ssd_sw = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'write') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

pt='hdd'
#confition for rbd_hdd
conditions_rbd_hdd_rr = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'randread') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

conditions_rbd_hdd_rw = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'randwrite') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

conditions_rbd_hdd_sr = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'read') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

conditions_rbd_hdd_sw = (
    (df['pool_type'] == pt) & 
    (df['disk_type'] == dt) & 
    (df['cache_size'] == cs) & 
    (df['workload'] == 'write') &
    (df['bs'] == bs) &
    (df['iodepth']==qd)
)

def extract_values(input_list):
    numeric_values = []
    converted_values = []

    for value_with_unit in input_list:
        match = re.match(r'(\d+\.\d+|\d+)(\w+)/s', value_with_unit)
        if match:
            numeric_value = float(match.group(1))
            unit = match.group(2)

            if unit == 'KiB':
                numeric_value /= 1024  # Convert KiB to MiB

            numeric_values.append(numeric_value)
            converted_values.append(f'{numeric_value}MiB/s')

    return numeric_values
```
